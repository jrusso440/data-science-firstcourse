# Density Estimation
What if we do not know how many clusters we have?
We can look for data points that are dense together and use that information to form cluster
One way is MeanShift
What is density?
Think about trying to find your favorite coffee shop (say Dunkin)  
  Higher probability in urban area as opposed to rural area
  Dunkin's has a higher probability or likelihood of being in an urban area
Density is a function to tell us the likelihood that a variable (i.e. Dunkin location) wil take on a given value
MeanShift returns a maxima (maximum values) of a density function given set of points that fit that function
If we have Dunkin data points, MeanShift will help us to figure out how likely you are to find the Dunkins at a location or how likely a location will have a Dunkins


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift
from sklearn.datasets import make_blobs
from mpl_toolkits.mplot3d import Axes3D

# Inport styleclass from matplotlib and use that to apply gcplot styling
from matplotlib import style

#Let's create a bunch of points around 3 centers in a 3d space space X has those
#points and we can ignore space y
centers = [[1,1,1],[5,5,5],[3,10,10]]
X,y = make_blobs(n_samples=100,centers=centers, cluster_std=2)

#perform clustering using MeanShift algorithm
ms=MeanShift()
ms.fit(X)

#"ms" hold the model; extract information clusters as
#represented by their centroids , along with their labels
centroids = ms.cluster_centers_
labels=ms.labels_

print(centroids)
print(labels)
```

```python
#find out how many clusters we created
n_clusters_ = len(np.unique(labels))
print("Number of estimated clusters:",n_clusters_)
```

```python
#define a colors array
colors = ['r','g','b','c','k','y','m']
```

```python
#loop to go through each data point, plotting it on the 3D space
#with a color picked from the above list
```

```python
#Let's do a 3D Plot
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
for i in range(len(X)):
  print("Coordinate: ",X[i],"Label: ",labels[i])
  ax.scatter(X[i][0],X[i][1],X[i][2],c=colors[labels[i]],marker='o')

ax.scatter(centroids[:,0],centroids[:,1],centroids[:,2],marker="x",s=150,linewidths=5,zorder=10)
plt.show()
```
